# Предел функции <a class='anchor' id='lims_func'>

### Задание 1 <a class='anchor' id='lims_func_1'>
    
Предложить пример функции, не имеющей предела в нуле и в бесконечностях.
    
$$f(x) = \frac{1}{\sin(x)}.$$

При $x \to \infty$ предел у этой функции не существует, при $x = 0$ имеются только односторонние пределы:

$$\lim_{x \to 0-} f(x) = -\infty \neq \infty = \lim_{x \to 0+} f(x).$$
    
Отобразим график для наглядности.

```python
x = np.linspace(-5, 5, 100)
y = 1.0 / np.sin(x)

plt.figure(figsize=(10, 10))
plt.axis([-5, 5, -10, 10])

plt.plot(x, y)
plt.grid()
plt.show()
```

$$f(x) = \text{sgn}(x) = \begin{cases}
  -1, x < 0 \\
   0, \; x = 0 \\
   1, \; x > 0
\end{cases}, \; x \in \mathbb{R}.$$
 
Эта функция определена в точке $x = 0$, но имеет только односторонние пределы в этой точке:
    
$$\lim_{x \to 0-} f(x) = -1 \neq 1 = \lim_{x \to 0+} f(x).$$

### Задание 3 <a class='anchor' id='lims_func_3'>
    
Исследовать функцию $f(x) = x^3 - x^2$ по плану:
    
* область задания и область значений;
* нули функции и их кратность;
* отрезки знакопостоянства;
* интервалы монотонности;
* четность функции;
* ограниченность;
* периодичность.

```python
x = np.linspace(-1, 2, 100)
y = np.power(x, 3) - np.power(x, 2)

plt.figure(figsize=(10, 10))

plt.plot(x, y)
plt.grid()
plt.show()
```

**Область определения** и **область значения** - вся числовая ось $\mathbb{R}$.

Для определения нулей функции приравняем её к нулю и решим уравнение:

$$x^3 - x^2 = x^2(x - 1) = 0.$$

Корни уравнения $x = 0$, $x = 1$.

Таким образом, получаем **нули функции**: 
$$x = 0 \; \text{кратности} \; 2,$$
$$x = 1 \; \text{кратности} \; 1.$$

Определим **отрезки знакопостоянства**, подставив значения из интервалов в функцию $f(x) = x^3 - x^2$:

1) $x > 1$: $f(2) = 4 > 0$;

2) $0 < x < 1$: $f(0.5) = -0.125 < 0$;

3) $x < 0$: $f(-1) = -2 < 0$.

Для наглядности отобразим схему с интервалами и полученными на них знаками.

Определим **интервалы монотонности**.

Для этого сначала найдем производную функции:

$$f'(x) = 3x^2 - 2x.$$

Далее определим нули производной, решив уравнение:

$$3x^2 - 2x = x(3x - 2) = 0.$$

Корни уравнения $x = 0$, $x = \frac{2}{3}.$

Определим знаки функции $g(x) = f'(x)$, подставив значения из интервалов в функцию $g(x) = 3x^2 - 2x$:

1) $x > \frac{2}{3}$: $g(1) = 1 > 0$;

2) $0 < x < \frac{2}{3}$: $g \left(\frac{1}{3} \right) = -\frac{1}{3} < 0$;

3) $x < 0$: $g(-1) = 5 > 0$.

Таким образом, функция возрастает на интервалах $(-\infty, 0)$, $\left(\frac{2}{3}, +\infty \right)$ и убывает на интервале $\left(0, \frac{2}{3} \right)$. Полученный результат совпадает с тем, что изображено на графике функции выше.

Для наглядности также отобразим схему с интервалами и знаками на них.

Исследуем функцию на четность:

$$f(x) = x^3 - x^2,$$

$$f(-x) = -x^3 - x^2.$$

Видно, что $f(-x)$ не равно $-f(x)$ и $f(x)$, то есть эта функция является **функцией общего вида**, что также видно на графике функции.

Очевидно, что функция **не ограничена сверху и снизу**.

Исследуем функцию на периодичность:

$$f(x + T) = (x + T)^3 - (x + T)^2 = x^3 + 3x^2T + 3xT^2 + T^3 - x^2 - 2xT - T^2 = $$

$$ = x^3 - x^2(1 - 3T) + xT(3T - 2) + (T^3 - T^2).$$

Очевидно, что полученное выражение равно функции $f(x) = x^3 - x^2$ только при $T = 0$, то есть эта функция **не периодическая**.

**Пример 1** <a class='anchor' id='lims_func_4_1'>
    
$$\lim_{x \to 0} \frac{3x^3 - 2x^2}{4x^2} = \lim_{x \to 0} \frac{x^2(3x - 2)}{4x^2} = \lim_{x \to 0} \frac{3x - 2}{4} = -\frac{1}{2}.$$

**Пример 2** <a class='anchor' id='lims_func_4_2'>
    
 $$\lim_{x \to 0} \frac{\sqrt{1 + x} - 1}{\sqrt[3]{1 + x} - 1} = \bigg| \; 1 + x = y \; \bigg| = \lim_{y \to 1} \frac{\sqrt{y} - 1}{\sqrt[3]{y} - 1} = \lim_{y \to 1} \frac{\left(\sqrt{y} - 1 \right) \left(\sqrt{y} + 1 \right) \left(\sqrt[3]{y^2} + \sqrt[3]{y} + 1 \right)}{ \left(\sqrt[3]{y} - 1 \right) \left(\sqrt[3]{y^2} + \sqrt[3]{y} + 1 \right) \left(\sqrt{y} + 1 \right)} = $$
    
$$ = \lim_{y \to 1} \frac{(y - 1) \left(\sqrt[3]{y^2} + \sqrt[3]{y} + 1 \right)}{(y - 1) \left(\sqrt{y} + 1 \right)} = \lim_{y \to 1} \frac{\sqrt[3]{y^2} + \sqrt[3]{y} + 1}{\sqrt{y} + 1} = \frac{3}{2}.$$
    
При решении были использованы следующие приемы:
1) была произведена замена $1 + x = y$;

2) числитель и знаменатель были умножены и разделены на сопряженный множитель:
    
$\sqrt{y} + 1$ - для числителя,
    
$\sqrt[3]{y^2} + \sqrt[3]{y} + 1$ - для знаменателя;
    
3) были использованы формулы:

$a^2 - b^2 = (a - b)(a + b)$ - для числителя: $\left(\sqrt{y} - 1 \right) \left(\sqrt{y} + 1 \right) = y - 1$,
    
$a^3 - b^3 = (a - b)(a^2 + ab + b^2)$ - для знаменятеля: $\left(\sqrt[3]{y} - 1 \right) \left(\sqrt[3]{y^2} + \sqrt[3]{y} + 1 \right) = y - 1$.

# Теоремы о пределах <a class='anchor' id='lims_theorem'>

### Задание 1 <a class='anchor' id='lims_theorem_1'>
    
Найти предел.

**Пример 1** <a class='anchor' id='lims_theorem_1_1'>
    
$$\lim_{x \to 0} \frac{\sin(2x)}{4x} = \lim_{x \to 0} \frac{\sin(2x)}{2x}\frac{2x}{4x} = \frac{1}{2} \lim_{x \to 0} \frac{\sin(2x)}{2x} = \frac{1}{2}.$$
    
При решении были использованы следующие приемы:
1) умножили и разделили на выражение на $2x$;

2) константа $\frac{1}{2}$ была вынесена за знак предела;

3) было использовано свойство первого замечательного предела:
    
$$\lim_{x \to 0} \frac{\sin(2x)}{2x} = 1.$$

**Пример 2** <a class='anchor' id='lims_theorem_1_2'>
    
$$\lim_{x \to 0} \frac{x}{\sin(x)} = 1.$$
    
Это является следствием первого замечательного предела.
