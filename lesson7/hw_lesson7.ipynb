### Задание 1 <a class='anchor' id='derivative_1'>
    
Найти длину $x$ и ширину $y$ прямоугольника при заданном периметре $P=144$ см, при которых данный прямоугольник имеет наибольшую площадь $S$.
    
Запишем функцию, для которой будем определять точки экстремума:
    
$$S(x, y) = x \cdot y,$$
    
где $S$ - площадь прямоугольника со сторонами $x$ и $y$. Это функция, зависящая от двух переменных.
    
Упростим задачу, используя условие, что периметр $P$ задан.
    
Формула для периметра прямоугольника:
    
$$P = x + y,$$
    
откуда можно выразить перменную $x$ или $y$. Выразим $x$:
    
$$x = P - y.$$

Подставим полученное выражение для $x$ в формулу для площади:
    
$$S(y) = (P - y) \cdot y = Py - y^2.$$
    
Получили функцию одной переменной.

Для определения точек экстремума найдем производную полученной функции:
    
$$S'(y) = P - 2y,$$
    
и решим уравнение:
    
$$P - 2y = 0.$$
    
Корень уравнения: 
    
$$y = \frac{P}{2}.$$
    
Проверим, что найденная точка является точкой максимума. 

Для этого определим знаки производной $S'(y)$ на интервалах слева и справа от найденной точки:
    
1) $S'(0) = P > 0$,
    
2) $S'(P) = -P < 0$.
    
Видно, что в точке $\frac{P}{2}$ производная $S'(y)$ меняет свой знак с плюса на минус, значит точка $y = \frac{P}{2}$ явлеятся точкой экстремума, а именно точкой максимума.
    
Зная выражение для $y$ можем определить чему равен $x$:
    
$$x = P - y = P - \frac{P}{2} = \frac{P}{2},$$
    
то есть при равных сторонах $x = y = \frac{P}{2}$ прямоугольник с заданным периметром $P$ будет иметь максимальную площадь.
    
Таким образом, получаем следующие значения для $x$, $y$:
    
$$x = y = \frac{P}{2} = \frac{144}{2} = 72.$$

### Задание 2 <a class='anchor' id='derivative_2'>
    
Найти экстремумы функций, если они есть.

**Пример 1** <a class='anchor' id='derivative_2_1'>
    
$$y = |2x|$$
    
Вычислим производную:

$$y' = \frac{2x}{|x|}.$$
    
Очевидно, что нет такой точки, при которой производная была бы равна нулю, но есть точка $x = 0$, в которой производная не существует. 
    
В точке $x = 0$ производная меняет знак с минуса на плюс, значит эта точка является точкой экстремума, а именно точкой минимума.
    
Для наглядности отобразим графики функции и её производной. Точка минимума обозначена зеленой вертикальной линией.

```python
x = np.linspace(-5, 5, 800)
y = np.abs(2 * x)
y_der = (2 * x) / np.abs(x)

plt.figure(figsize=(10, 10))
plt.axis([-5, 5, -3, 10])

plt.plot(x, y)
plt.plot(x, y_der)
plt.vlines(0, -3, 10, linestyles ='dashed', color='green')

plt.grid()
plt.show()
```

**Пример 2** <a class='anchor' id='derivative_2_2'>
    
$$y = x^3$$
    
Вычислим производную:

$$y' = 3x^2,$$

приравняем полученное выражение к нулю и решим уравнение:

$$3x^2 = 0 \Rightarrow x = 0.$$

Корень уравнения: $x = 0$ кратности 2.
    
Очевидно, что в точке $x = 0$ функция не меняет знак, значит эта критическая точка не является точкой экстремума. Это точка перегиба.
    
Для наглядности отобразим графики функции и её производной. Точка перегиба обозначена зеленой вертикальной линией.

```python
x = np.linspace(-5, 5, 100)
y = np.power(x, 3)
y_der = 3 * np.power(x, 2)

plt.figure(figsize=(10, 10))
plt.axis([-2, 2, -5, 5])

plt.plot(x, y, label='x^3')
plt.plot(x, y_der, label='3x^2')
plt.vlines(0, -10, 10, linestyles ='dashed', color='green')

plt.grid()
plt.show()
```

**Пример 3** <a class='anchor' id='derivative_2_3'>
    
$$y = e^{3x}$$
    
Вычислим производную:
    
$$y' = 3e^{3x}.$$
    
Очевидно, что у этой функции нет точек, в которых её производная равна нулю или не существует, то есть эта функция не имеет критических точек.
    
Ниже изображен график функции и её производной.

```python
x = np.linspace(-5, 5, 100)
y = np.exp(3 * x)
y_der = 3 * np.exp(3 * x)

plt.figure(figsize=(10, 10))
plt.axis([-2, 2, -2, 10])

plt.plot(x, y, label='e^(3x)')
plt.plot(x, y_der, label='3e^(3x)')

plt.legend()
plt.grid()
plt.show()
```

**Пример 4** <a class='anchor' id='derivative_2_4'>
    
$$y = x^3-5x$$

Вычислим производную функции:

$$y' = 3x^2 - 5,$$

приравняем полученное выражение к нулю и решим уравнение:

$$3x^2 - 5 = 0.$$

$$3x^2 = 5,$$

$$x^2 = \frac{5}{3} \Rightarrow x_{1, 2} = \pm \sqrt{\frac{5}{3}},$$

то есть корни уравнения: $x_{1, 2} = \pm \sqrt{\frac{5}{3}}$.

Определим знаки производной, подставив значения из интервалов в функцию $y' = 3x^2 - 5$:

1) $x < -\sqrt{\frac{5}{3}}$: $y'\left(-\sqrt{5} \right) = 10 > 0$;

2) $-\sqrt{\frac{5}{3}} < x < \sqrt{\frac{5}{3}}$: $y'(0) = -5 < 0$;

3) $x > \sqrt{\frac{5}{3}}$: $y'\left(\sqrt{5} \right) = 10 > 0$.

Результат подстановки также изображен на схеме.

<img src="task4.png">

Из схемы видно, что на этих точках функция меняет знак, значит точки $x_{1, 2}$ - точки экстремума,

причем $x_1 = -\sqrt{\frac{5}{3}}$ - точка максимума, $x_2 = \sqrt{\frac{5}{3}}$ - точка минимума.

Таким образом, функция возрастает на интервалах $\left(-\infty, -\sqrt{\frac{5}{3}} \right)$, $\left(\sqrt{\frac{5}{3}}, +\infty \right)$ и убывает на интервале $\left(-\sqrt{\frac{5}{3}}, \sqrt{\frac{5}{3}} \right)$.

Для наглядности отобразим графики функции и её производной. Точки экстремума изображены вертикальными линиями.

```python
x = np.linspace(-5, 5, 100)
y = np.power(x, 3) - 5 * x
y_der = 3 * np.power(x, 2) - 5

plt.figure(figsize=(10, 10))
plt.axis([-5, 5, -10, 10])

x1 = -np.sqrt(5 / 3)
x2 = np.sqrt(5 / 3)

plt.plot(x, y, label = 'x^3 - 5x')
plt.plot(x, y_der, label='3x^2 - 5')
plt.vlines(x1, -10, 10, linestyles ='dashed', color='green')
plt.vlines(x2, -10, 10, linestyles ='dashed', color='green')

plt.legend()
plt.grid()
plt.show()
```
