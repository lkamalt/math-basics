### Задание 1 <a class='anchor' id='series_1'>
    
Исследовать ряд на сходимость, используя признак д'Аламбера

$$\sum_{n=1}^{\infty} \frac{n^n}{(n!)^2}.$$
    
Признак д'Аламбера
    
$$\lim_{n \to \infty} \frac{a_{n + 1}}{a_n} = q,$$
    
если $q < 1$ - ряд сходится, если $q > 1$ - расходится.
    
Вычислим предел
    
$$q = \lim_{n \to \infty} \frac{a_{n + 1}}{a_n} = \lim_{n \to \infty} \frac{(n + 1)^{n + 1}}{((n + 1)!)^2} \frac{(n!)^2}{n^n} = \lim_{n \to \infty} \frac{(n + 1)^{n + 1}}{(n!)^2(n + 1)^2} \frac{(n!)^2}{n^n} = \lim_{n \to \infty} \frac{(n + 1)^{n + 1}}{(n + 1)^2 n^n} =$$
    
$$= \lim_{n \to \infty} \frac{n^{n + 1} \left( 1 + \frac{1}{n} \right)^{n + 1}}{n^{n + 2} \left( 1 + \frac{1}{n} \right)^2} = \lim_{n \to \infty} \frac{1}{n \left( 1 + \frac{1}{n} \right)^2} \left( 1 + \frac{1}{n} \right)^{n + 1} = 0 \cdot e = 0.$$
    
Получаем

$$q = 0 < 1,$$

то есть ряд **сходится**.

### Задание 2 <a class='anchor' id='series_2'>
    
Исследовать ряд на сходимость, используя радикальный признак Коши
    
$$\sum_{n=1}^{\infty} \frac{n}{2^n}.$$
    
Радикальный признак Коши
    
$$\lim_{n \to \infty} \sqrt[n]{a_n} = q,$$
    
если $q < 1$, то ряд сходится, если $q > 1$, то расходится, если $q = 1$, то признак в этом случае не работает.
    
Вычислим предел
    
$$q = \lim_{n \to \infty} \sqrt[n]{a_n} = \lim_{n \to \infty} \sqrt[n]{\frac{n}{2^n}} = \lim_{n \to \infty} \frac{\sqrt[n]{n}}{2} = \frac{1}{2} \lim_{n \to \infty} \sqrt[n]{n} = \frac{1}{2} \cdot 1 = \frac{1}{2}.$$
    
При вычислении был использован предел:
    
$$\lim_{n \to \infty} \sqrt[n]{n} = 1.$$
    
Таким образом,
    
$$q = \frac{1}{2} < 1,$$
    
то есть ряд **сходится**.

### Задание 3 <a class='anchor' id='series_3'>
    
Исследовать ряд на сходимость, используя признак Лейбница
    
$$\sum_{n=1}^{\infty} \frac{(-1)^n}{n + \ln{n}}.$$
    
Признак Лейбница:

если

1. $|a_n| \ge |a_{n + 1}|$, начиная с некоторого номера $n \ge N$,
    
2. $\lim_{n \to \infty} a_n = 0$,
    
то ряд сходится.

Если при этом сходится и ряд из модулей, то исходный ряд сходится абсолютно, иначе - сходится условно.
    
Проверим первое условие, для этого рассмотрим разность слагаемых:
    
$$|a_n| - |a_{n + 1}| = \frac{1}{n + \ln{n}} - \frac{1}{n + 1 + \ln(n + 1)} > 0, \; \forall n \geq 1. $$
    
Очевидно, что разница модулей при $\forall n \geq 1$ является положительным числом, то есть первое слагаемое больше, чем второе слагаемое, потому что логарифм - монотонно возрастающая функция, поэтому знаменатель второй дроби оказывается больше, чем знаменатель первой дроби. То есть первое условие признака выполняется.
    
Проверим второе условие, для этого вычислим предел:
 
$$\lim_{n \to \infty} a_n = \lim_{n \to \infty} \frac{(-1)^n}{n + \ln{n}} = \lim_{n \to \infty} \frac{(-1)^n}{n \left( 1 + \frac{\ln{n}}{n} \right)} = 0,$$
    
видно, что второе условие тоже выполняется.
    
Таким образом, исходный ряд сходится. Теперь рассмотрим ряд из модулей.
    
$$\sum_{n=1}^{\infty} \frac{1}{n + \ln n}.$$
    
Этот ряд расходится по интегральному признаку Коши, так как расходится несобственный интеграл
    
$$\int_1^{+\infty} \frac{dx}{x + \ln x}.$$
    
Получили, что исходный ряд **сходится условно**.

### Задание 4 <a class='anchor' id='series_4'>
    
Исследовать ряд на сходимость, используя признак Раабе
    
$$\sum_{n=1}^{\infty} \frac{3^n}{2^n}.$$
    
Признак Раабе
    
$$\lim_{n \to \infty} \left( n \left( \frac{a_n}{a_{n + 1}} - 1 \right) \right) = D,$$
    
если $D < 1$, то ряд расходится, если $D > 1$, то сходится, при $D = 1$ признак не дает ответа.
    
Вычислим предел
    
$$D = \lim_{n \to \infty} \left( n \left( \frac{a_n}{a_{n + 1}} - 1 \right) \right) = \lim_{n \to \infty} \left( n \left( \frac{3^n}{2^n}\frac{2^{n + 1}}{3^{n + 1}} - 1 \right) \right) = \lim_{n \to \infty} \left( n \left( \frac{2}{3} - 1 \right) \right) = -\frac{1}{3} \lim_{n \to \infty} n = -\infty.$$
    
Таким образом,
    
$$D = -\infty < 1,$$
    
то есть ряд **расходится**.

### Задание 5 <a class='anchor' id='series_5'>
    
Разложить функцию по Тейлору в единице
    
$$f(x) = \ln(16x^2).$$
    
Ряд Тейлора функции $f(x)$ в точке $a$:
    
$$f(x) = \sum_{n = 0}^{\infty}\frac{f^{(n)}a}{n!}(x - a)^n.$$
    
Запишем производные заданной функции (первые 7):
    
$$f^{(1)}(x) = \frac{16 \cdot 2x}{16x^2} = \frac{2}{x},$$
    
$$f^{(2)}(x) = -\frac{2}{x^2},$$
    
$$f^{(3)}(x) = \frac{4}{x^3},$$
    
$$f^{(4)}(x) = -\frac{12}{x^4},$$
    
$$f^{(5)}(x) = \frac{48}{x^5},$$
    
$$f^{(6)}(x) = -\frac{240}{x^6},$$
    
$$f^{(7)}(x) = \frac{1440}{x^7}.$$
    
Запишем первые 7 слагаемых ряда Тейлора в точке $a = 1$:

$$f(1) = \ln(16),$$
    
$$f^{(1)}(1)(x - 1) = 2(x - 1),$$
    
$$\frac{f^{(2)}(1)}{2!}(x - 1)^2 = -(x - 1)^2,$$
    
$$\frac{f^{(3)}(1)}{3!}(x - 1)^3 = \frac{2}{3}(x - 1)^3,$$
    
$$\frac{f^{(4)}(1)}{4!}(x - 1)^4 = -\frac{1}{2}(x - 1)^4,$$
    
$$\frac{f^{(5)}(1)}{5!}(x - 1)^5 = \frac{2}{5}(x - 1)^5,$$
    
$$\frac{f^{(6)}(1)}{6!}(x - 1)^6 = -\frac{1}{3}(x - 1)^6,$$
    
$$\frac{f^{(7)}(1)}{7!}(x - 1)^7 = \frac{2}{7}(x - 1)^7.$$
    
Тогда ряд Тейлора функции $f(x) = \ln(16x^2)$ в точке $a = 1$ в общем виде можно записать так:
    
$$f(x) = \ln(16) - \sum_{n = 1}^{\infty}\frac{1}{n}(x - 1)^{2n} + \sum_{n = 1}^{\infty} \frac{2}{2n - 1}(x - 1)^{2n - 1}.$$
