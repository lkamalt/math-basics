### Задание 1 <a class='anchor' id='derivative_1'>
    
Исследовать функцию на условный экстремум.

**Пример 1** <a class='anchor' id='derivative_1_1'>
    
$$U = 3 - 8x + 6y, \; \text{если} \; x^2 + y^2 = 36 $$
    
Запишем функцию Лагранжа:

$$L(x, y, \lambda) = U(x, y) + \lambda \varphi(x, y) = 3 - 8x + 6y + \lambda(x^2 + y^2 - 36).$$
    
Необходимое условие экстремума: 
    
$$\begin{cases}
  L'_x = -8 + 2\lambda x = 0, \\
  L'_y = 6 + 2\lambda y = 0, \\
  L'_{\lambda} = x^2 + y^2 - 36 = 0.
\end{cases}$$
 
Выразим из первого уравнения $x$:

$$x = 4\lambda,$$

из второго уравнения выразим $y$:
    
$$y = -3\lambda,$$

подставим полученные выражения у третье уравнение, получим уравнение на $\lambda$:
    
$$16\lambda^2 + 9\lambda^2 - 36 = 25\lambda^2 - 36 = 0.$$
    
Из полученного уравнения найдем $\lambda$:
    
$$\lambda = \pm \frac{6}{5}.$$

Таким образом, подставляя значения для $\lambda$ в выражения для $x$ и $y$, получаем две стационарные точки:
    
$$M_1 = \left( \frac{24}{5}; -\frac{18}{5} \right),$$
    
$$M_2 = \left( -\frac{24}{5}; \frac{18}{5} \right).$$
    
Исследуем эти точки на экстремум.

Найдем вторые производные функции Лагранжа:
    
$$L''_{xx} = 2\lambda, \; L''_{yy} = 2\lambda, \; L''_{\lambda \lambda} = 0,$$
    
$$L''_{xy} = 0, \; L''_{x\lambda} = 2x, \; L''_{y\lambda} = 2y.$$
    
Вычислим гессиан в общем виде:
    
$$
\Delta = \begin{vmatrix}
    L''_{\lambda \lambda} & L''_{\lambda x} & L''_{\lambda y} \\
    L''_{x\lambda} & L''_{xx} & L''_{xy} \\
    L''_{y\lambda} & L''_{yx} & L''_{yy}
\end{vmatrix} = 
\begin{vmatrix}
    0 & 2x & 2y \\
    2x & 2\lambda & 0 \\
    2y & 0 & 2\lambda
\end{vmatrix} = -2x 
\begin{vmatrix}
    2x & 0 \\
    2y & 2\lambda \\
\end{vmatrix} + 2y
\begin{vmatrix}
    2x & 2\lambda \\
    2y & 0 \\
\end{vmatrix} = 
$$

$$= -2x \cdot 4x\lambda - 2y \cdot 4y\lambda = -8x^2\lambda - 8y^2\lambda = -8\lambda(x^2 + y^2).$$
    
Вычислим гессиан в первой точке $M_1$ при $\lambda = \frac{6}{5}$:
    
$$\Delta_1 = -8 \frac{6}{5}(x^2 + y^2) = -\frac{48}{5}(x^2 + y^2) < 0.$$
    
Получили, что в стационарной точке $M_1 = \left( \frac{24}{5}; -\frac{18}{5} \right)$ определитель матрицы $\Delta_1 < 0$, значит для функции $U$ эта точка является **точкой минимума**.
    
Вычислим гессиан во второй точке $M_2$ при $\lambda = -\frac{6}{5}$:
    
$$\Delta_2 = 8 \frac{6}{5}(x^2 + y^2) = \frac{48}{5}(x^2 + y^2) > 0.$$
    
Получили, что во второй стационарной точке $M_2 = \left( -\frac{24}{5}; \frac{18}{5} \right)$ определитель матрицы $\Delta_2 > 0$, значит для функции $U$ эта точка является **точкой максимума**.

**Пример 2** <a class='anchor' id='derivative_1_2'>
    
$$U = 2x^2 + 12xy + 32y^2 + 15, \; \text{если} \; x^2 + 16y^2 = 64 $$
    
Запишем функцию Лагранжа:

$$L(x, y, \lambda) = U(x, y) + \lambda \varphi(x, y) = 2x^2 + 12xy + 32y^2 + 15 + \lambda(x^2 + 16y^2 - 64).$$
    
Необходимое условие экстремума: 
    
$$\begin{cases}
  L'_x = 4x + 12y + 2x\lambda = 0, \\
  L'_y = 12x + 64y + 32y\lambda = 0, \\
  L'_{\lambda} = x^2 + 16y^2 - 64 = 0.
\end{cases}$$
    
Перепишем систему:

$$\begin{cases}
  L'_x = 6y + x(2 + \lambda) = 0, \\
  L'_y = 3x + 8y(2 + \lambda) = 0, \\
  L'_{\lambda} = x^2 + 16y^2 - 64 = 0.
\end{cases}$$
    
Обозначим
    
$$a = 2 + \lambda,$$
    
тогда система будет выглядеть следующим образом:
    
$$\begin{cases}
  L'_x = 6y + ax = 0, \\
  L'_y = 3x + 8ay = 0, \\
  L'_{\lambda} = x^2 + 16y^2 - 64 = 0.
\end{cases}$$
    
Выразим $x$ из первого уравнения:

$$x = -\frac{6y}{a},$$

и подставим во второе и третье

$$\begin{cases}
  -\frac{18y}{a} + 8ay = 0, \\
  \frac{36y^2}{a^2} + 16y^2 - 64 = 0, \\
\end{cases} \Rightarrow
\begin{cases}
  -18y + 8a^2y = 0, \\
  36y^2 + 16a^2y^2 - 64a^2 = 0. \\
\end{cases}
$$
    
Из первого уравнения полученной системы выразим, чему равно $a$:
    
$$a^2 = \frac{18}{8} = \frac{9}{4} \; \Rightarrow \; a = \pm \frac{3}{2}.$$
    
Далее умножим первое уравнение на $2y$:
    
$$
\begin{cases}
  -36y^2 + 16a^2y^2 = 0, \\
  36y^2 + 16a^2y^2 - 64a^2 = 0. \\
\end{cases}
$$
    
Сложим первое уравнение со вторым:
    
$$32y^2a^2 - 64a^2 = 0,$$

откуда получаем выражение для $y$:
    
$$y^2 = \frac{64}{32} = 2 \; \Rightarrow \; y = \pm\sqrt{2}.$$
    
Таким образом, для разных значений $a$ и $y$ получаем 4 точки:
    
1) $a = \frac{3}{2}, \; y = \sqrt{2} \; \Rightarrow \; x = -4\sqrt{2}$

$$M_1 = \left( -4\sqrt{2}; \sqrt{2} \right),$$

2) $a = -\frac{3}{2}, \; y = \sqrt{2} \; \Rightarrow \; x = 4\sqrt{2}$

$$M_2 = \left( 4\sqrt{2}; \sqrt{2} \right),$$
    
3) $a = \frac{3}{2}, \; y = -\sqrt{2} \; \Rightarrow \; x = 4\sqrt{2}$

$$M_3 = \left( 4\sqrt{2}; -\sqrt{2} \right),$$
    
4) $a = -\frac{3}{2}, \; y = -\sqrt{2} \; \Rightarrow \; x = -4\sqrt{2}$

$$M_4 = \left( -4\sqrt{2}; -\sqrt{2} \right).$$

Исследуем эти точки на экстремум.

Найдем вторые производные функции Лагранжа:
    
$$L''_{xx} = 4 + 2\lambda, \; L''_{yy} = 64 + 32\lambda, \; L''_{\lambda \lambda} = 0,$$
    
$$L''_{xy} = 12, \; L''_{x\lambda} = 2x, \; L''_{y\lambda} = 32y.$$
    
Вычислим гессиан в общем виде с учетом того, что $a = 2 + \lambda$:
    
$$
\Delta = \begin{vmatrix}
    L''_{\lambda \lambda} & L''_{\lambda x} & L''_{\lambda y} \\
    L''_{x\lambda} & L''_{xx} & L''_{xy} \\
    L''_{y\lambda} & L''_{yx} & L''_{yy}
\end{vmatrix} = 
\begin{vmatrix}
    0 & 2x & 32y \\
    2x & 4 + 2\lambda & 12 \\
    32y & 12 & 64 + 32\lambda
\end{vmatrix} = 8
\begin{vmatrix}
    0 & x & 16y \\
    x & a & 6 \\
    16y & 6 & 16a
\end{vmatrix}    
= -8x 
\begin{vmatrix}
    x & 6 \\
    16y & 16a \\
\end{vmatrix} + 128y
\begin{vmatrix}
    x & a \\
    16y & 6 \\
\end{vmatrix} =$$

$$-8x(16ax - 96y) + 128y(6x - 16ay) = -128ax^2 + 768xy + 768xy - 2048ay^2 = 1536xy - 128a(x^2 + 16y^2).$$
    
Получили гессиан
    
$$\Delta = 1536xy - 128a(x^2 + 16y^2).$$
    
1) Вычислим гессиан для первой точки $M_1\left(- 4\sqrt{2}; \sqrt{2}\right)$ при $a = \frac{3}{2}$:

$$\Delta_1 = -1536 \cdot 4\sqrt{2} \cdot \sqrt{2} - 128 \frac{3}{2}(x^2 + 16y^2) < 0.$$

Точка $M_1\left(- 4\sqrt{2}; \sqrt{2}\right)$ - точка минимума.
    
2) Вычислим гессиан для второй точки $M_2\left(4\sqrt{2}; \sqrt{2}\right)$ при $a = -\frac{3}{2}$:

$$\Delta_2 = 1536 \cdot 4\sqrt{2} \cdot \sqrt{2} + 128 \frac{3}{2}(x^2 + 16y^2) > 0.$$

Точка $M_2\left(4\sqrt{2}; \sqrt{2}\right)$ - точка максимума.

3) Вычислим гессиан для третьей точки $M_3\left(4\sqrt{2}; -\sqrt{2}\right)$ при $a = \frac{3}{2}$:

$$\Delta_3 = -1536 \cdot 4\sqrt{2} \cdot \sqrt{2} - 128 \frac{3}{2}(x^2 + 16y^2) < 0.$$

Точка $M_3\left(4\sqrt{2}; -\sqrt{2}\right)$ - точка минимума.

4) Вычислим гессиан для четвертой точки $M_4\left(-4\sqrt{2}; -\sqrt{2}\right)$ при $a = -\frac{3}{2}$:

$$\Delta_4 = 1536 \cdot 4\sqrt{2} \cdot \sqrt{2} + 128 \frac{3}{2}(x^2 + 16y^2) > 0.$$

Точка $M_4\left(-4\sqrt{2}; -\sqrt{2}\right)$ - точка максимума.

Таким образом, $M_1$, $M_2$, $M_3$, $M_4$ - точки экстремума функции $U = 2x^2 + 12xy + 32y^2 + 15$, а именно

$M_1\left(- 4\sqrt{2}; \sqrt{2}\right)$ - точка минимума,
    
$M_2\left(4\sqrt{2}; \sqrt{2}\right)$ - точка максимума,

$M_3\left(4\sqrt{2}; -\sqrt{2}\right)$ - точка минимума,

$M_4\left(-4\sqrt{2}; -\sqrt{2}\right)$ - точка максимума.

### Задание 2 <a class='anchor' id='derivative_2'>
    
Найти производную функции $U = x^2 + y^2 + z^2$ по направлению вектора $\vec{c} = (-9, 8, -12)$ в точке $M(8; -12; 9)$.
    
Производная по направлению вектора $\vec{c}$ в точке $M$ вычисляется по формуле:
    
$$\frac{\partial{U}}{\partial{c}} = \left. \frac{\partial{U}}{\partial{x}} \right|_M \cos\alpha + \left. \frac{\partial{U}}{\partial{y}} \right|_M \cos\beta + \left. \frac{\partial{U}}{\partial{z}} \right|_M \cos\gamma,$$
    
где $\cos\alpha, \cos\beta, \cos\gamma$ - координаты вектора единичной длины, определяющего направление вектора $\vec{c}$.
    
Вычислим частные производные:
    
$$\frac{\partial{U}}{\partial{x}} = 2x,$$
    
$$\frac{\partial{U}}{\partial{y}} = 2y,$$
    
$$\frac{\partial{U}}{\partial{z}} = 2z.$$
    
Вычислим значения частных производных в точке $M(8; -12; 9)$:
    
$$\left. \frac{\partial{U}}{\partial{x}} \right|_{(8; -12; 9)} = 2 \cdot 8 = 16,$$
    
$$\left. \frac{\partial{U}}{\partial{y}} \right|_{(8; -12; 9)} = -2 \cdot 12 = -24,$$
    
$$\left. \frac{\partial{U}}{\partial{z}} \right|_{(8; -12; 9)} = 2 \cdot 9 = 18.$$
    
Длина вектора $\vec{c}$:
    
$$|\vec{c}| = \sqrt{9^2 + 8^2 + 12^2} = \sqrt{289} = 17,$$
    
видно, что вектор $\vec{c}$ не единичной длины, поэтому нормируем вектор:
    
$$\vec{c_0} = \left( -\frac{9}{17}, \frac{8}{17}, -\frac{12}{17} \right),$$

Таким образом, производная функции $U$ по направлению вектора $\vec{c}$ в точке $M$ имеет вид:
    
$$\frac{\partial{U}}{\partial{c}} = -16 \cdot \frac{9}{17} - 24 \cdot \frac{8}{17} - 18 \cdot \frac{12}{17} = -\frac{144}{17} - \frac{192}{17} - \frac{216}{17} = -\frac{552}{17}.$$

### Задание 3 <a class='anchor' id='derivative_3'>
    
Найти производную функции $U = e^{x^2 + y^2 + z^2}$ по направлению вектора $\vec{d} = (4, -13, -16)$ в точке $L(-16; 4; -13)$.
    
Производная по направлению вектора $\vec{d}$ в точке $L$ вычисляется по формуле:
    
$$\frac{\partial{U}}{\partial{d}} = \left. \frac{\partial{U}}{\partial{x}} \right|_L \cos\alpha + \left. \frac{\partial{U}}{\partial{y}} \right|_L \cos\beta + \left. \frac{\partial{U}}{\partial{z}} \right|_L \cos\gamma,$$
    
где $\cos\alpha, \cos\beta, \cos\gamma$ - координаты вектора единичной длины, определяющего направление вектора $\vec{d}$.
    
Вычислим частные производные:
    
$$\frac{\partial{U}}{\partial{x}} = 2x e^{x^2 + y^2 + z^2},$$
    
$$\frac{\partial{U}}{\partial{y}} = 2y e^{x^2 + y^2 + z^2},$$
    
$$\frac{\partial{U}}{\partial{z}} = 2z e^{x^2 + y^2 + z^2}.$$
    
Вычислим значения частных производных в точке $L(-16; 4; -13)$:
    
$$\left. \frac{\partial{U}}{\partial{x}} \right|_{(-16; 4; -13)} = -32 e^{441},$$
    
$$\left. \frac{\partial{U}}{\partial{y}} \right|_{(-16; 4; -13)} = 8 e^{441},$$
    
$$\left. \frac{\partial{U}}{\partial{z}} \right|_{(-16; 4; -13)} = -26 e^{441}.$$
    
Длина вектора $\vec{d}$:
    
$$|\vec{d}| = \sqrt{4^2 + 13^2 + 16^2} = \sqrt{441} = 21,$$
    
видно, что вектор $\vec{d}$ не единичной длины, поэтому нормируем вектор:
    
$$\vec{d_0} = \left( \frac{4}{21}, -\frac{13}{21}, -\frac{16}{21} \right),$$

Таким образом, производная функции $U$ по направлению вектора $\vec{d}$ в точке $L$ имеет вид:
    
$$\frac{\partial{U}}{\partial{d}} = -32 e^{441} \cdot \frac{4}{21} - 8 e^{441} \cdot \frac{13}{21} + 26 e^{441} \cdot \frac{16}{21} = e^{441} \left( -\frac{128}{21} - \frac{104}{21} + \frac{416}{21} \right) = e^{441}\frac{184}{21}.$$
